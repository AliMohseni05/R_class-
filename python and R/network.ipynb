```python
#imfomation
a=["..................imfomation of graph....................."]

non=g.number_of_nodes()
noe=g.number_of_edges()
leng=len(g)
sizg=g.size()
listn=list(g.nodes)
liste=list(g.edges)
sgl1=list(g.adj[1]) # or list(G.neighbors(1))
g_degree=dict(g.degree)[2] # sum of links

#shortest path
sht_path=nx.shortest_path(g,2,4)
#centrality
dc=nx.degree_centrality(g)
cc=nx.closeness_centrality(g)
bc=nx.betweenness_centrality(g)
density=nx.density(g) #to complely graght
#eigen
diamter=nx.diameter(g) #long path 
#sub graph 
eulerian_path=nx.eulerian_path(g) #go to all and come back here agian 
cliques=list(nx.find_cliques(g))
bridges=list(nx.bridges(g))
lb=list(nx.local_bridges(g)) # if we remove it we must go longer path 
conc=list(nx.connected_components(g))

#print(nx.k_core(2)) #sube note at less have conneteted 
#print(nx.weakly_connected_components(g))    #for direck graph
#print(nx.strongly_connected_components(g)) #for direck graph
```
